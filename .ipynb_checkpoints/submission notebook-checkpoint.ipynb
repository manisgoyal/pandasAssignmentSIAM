### b. “Make” v/s “Doors”

```python
Make_Door_df = DataFrame3[DataFrame3['Make'].notna() & DataFrame3['Doors'].notna()]
Make_Door_df_BMW = Make_Door_df.where(Make_Door_df.Make=='BMW').groupby('Doors').count()['Make']
Make_Door_df_Toyota = Make_Door_df.where(Make_Door_df.Make=='Toyota').groupby('Doors').count()['Make']
Make_Door_df_Nissan = Make_Door_df.where(Make_Door_df.Make=='Nissan').groupby('Doors').count()['Make']
Make_Door_df_Honda = Make_Door_df.where(Make_Door_df.Make=='Honda').groupby('Doors').count()['Make']
Make_Door_df
```

```python
Make_Door_df_BMW = pd.DataFrame(Make_Door_df_BMW)
Make_Door_df_BMW = Make_Door_df_BMW.transpose()
Make_Door_df_BMW.index = ['BMW']
Make_Door_df_BMW
```

```python
Make_Door_df_Nissan = pd.DataFrame(Make_Door_df_Nissan)
Make_Door_df_Nissan = Make_Door_df_Nissan.transpose()
Make_Door_df_Nissan.index = ['Nissan']
Make_Door_df_Nissan
```

```python
Make_Door_df_Toyota = pd.DataFrame(Make_Door_df_Toyota)
Make_Door_df_Toyota = Make_Door_df_Toyota.transpose()
Make_Door_df_Toyota.index = ['Toyota']
Make_Door_df_Toyota
```

```python
Make_Door_df_Honda = pd.DataFrame(Make_Door_df_Honda)
Make_Door_df_Honda = Make_Door_df_Honda.transpose()
Make_Door_df_Honda.index = ['Honda']
Make_Door_df_Honda
```

```python
processed_Make_doors = pd.concat([Make_Door_df_BMW,Make_Door_df_Honda,Make_Door_df_Nissan,Make_Door_df_Toyota])
processed_Make_doors = processed_Make_doors.fillna(0)
processed_Make_doors
```

```python
makedoor = processed_Make_doors.plot.bar()
makedoor.set_title("'Make' V/S 'Doors'")
makedoor.set_ylabel("Number of Cars")
makedoor.set_xlabel('Car Manufacturers')
```

## 5) Create a dataframe using multiple functions with the following columns, where ‘n’ is the row index:
    a. Column 1: Sum of natural numbers up to nth row
    b. Column 2: Sum of the squares of natural numbers up to nth row
    c. Column 3: Sum of the cubes of natural numbers up to nth row
    d. Column 4: Factorial of the nth number
Example:


```python
math_df = pd.DataFrame([[1,1,1,1],
                       [3,5,9,2],
                       [6,14,36,6],
                       [10,30,100,24]], columns=['sum_num','sum_sq','sum_cube','factorial'])
math_df.index +=1

def amendTable(n):
    if valueCheck(n-1):
        math_df.loc[n] = [n+getCol1(n-1),n**2+getCol2(n-1),n**3+getCol3(n-1),n*getCol4(n-1)]
    else:
        amendTable(n-1)

def valueCheck(n):
    return 1 if n in math_df.index else 0

def getCol1(n):
    if valueCheck(n):
        return math_df.loc[n].at['sum_num']
    else:
        amendTable(n)
        return getCol1(n)

def getCol2(n):
    if valueCheck(n):
        return math_df.loc[n].at['sum_sq']
    else:
        amendTable(n)
        return getCol2(n)

def getCol3(n):
    if valueCheck(n):
        return math_df.loc[n].at['sum_cube']
    else:
        amendTable(n)
        return getCol3(n)

def getCol4(n):
    if valueCheck(n):
        return math_df.loc[n].at['factorial']
    else:
        amendTable(n)
        return getCol4(n)
```
